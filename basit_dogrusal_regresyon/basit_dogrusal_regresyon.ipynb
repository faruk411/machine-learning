```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
veri = pd.read_csv('Advertising.csv')
veri2 = veri.drop(veri.columns[0],axis=1)
veri2.head()
```

```python
veri2.info()
```

```python
sns.jointplot(x= 'TV', y='sales', data= veri2, kind='reg');
```

```python
X = veri2[['TV']] # bamsz deiken
y = veri2[['sales']] # baml deiken
```

```python
reg = LinearRegression()
model = reg.fit(X,y)
```

```python
model
```

```python
# beta 0 kat says
model.intercept_
```

```python
# beta 1 kat says
model.coef_
```

```python
# r kare
model.score(X,y) # baml deikenlerin , bamsz deikenlerce aklnama yzdesi
```

```python
g = sns.regplot(x=veri2['TV'], y=veri2['sales'], ci=None, scatter_kws={'color':'r','s':9})
g.set_title('Model denkelmi = 7.03 +TV*0.5')
g.set_ylabel('Sati sayisi')
g.set_xlabel('TV harcamalari')
plt.xlim(-10,310)
plt.ylim(bottom=0)
plt.show()
```
